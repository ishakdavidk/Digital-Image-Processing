```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def transform(image, original_pts, desired_pts, title_list):
    for i in range(4):
        cv2.circle(image, tuple(original_pts[i]), 5, (0, 0, 255), -1)
    
    matrix = cv2.getPerspectiveTransform(original_pts, desired_pts)
    result = cv2.warpPerspective(image, matrix, (331, 191))

    res_show = []
    res_show.append(image)
    res_show.append(result)
    
    fig = plt.figure(figsize=(13, 9))

    for i in range(2):
        fig.add_subplot(1,2,i+1)
        plt.title(title_list[i])
        plt.imshow(cv2.cvtColor(res_show[i], cv2.COLOR_BGR2RGB))
        plt.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
img = cv2.imread("dsu1.jpg")

title_list = ['DSU 1 : Original Image','DSU 1 : Perspective Transform Result']

original_pts = np.float32([[238, 268], [569, 233], [237, 459], [565, 494]])
desired_pts = np.float32([[0, 0], [331, 0], [0, 191], [331, 191]])
transform(img, original_pts, desired_pts, title_list)
```
