# Basic Operations on Images

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = (10,5)
```

```python
img=cv2.imread('messi.jpg')
```

```python
# access a pixel RGB value by its row and column coordinates 
px=img[100,100]
print(px)
```

```python
# accessing only blue pixel
blue=img[100,100,0]
print(blue)
```

```python
# modify pixel values
img[100,100]=[255,255,255]
print(img[100,100])
```

```python
# accessing RED value
img.item(10,10,2)
```

```python
# modifying RED value
img.itemset((10,10,2),100)
img.item(10,10,2)
```

```python
plt.imshow(img)
```

```python
print(img.shape)
```

```python
print(img.size)
```

```python
print(img.dtype)
```

## Image ROI

```python
ball=img[280:340,330:390]
img[273:333,100:160]=ball
```

```python
plt.imshow(img)
```

### Splitting and Merging Image Channels

```python
b,g,r=cv2.split(img)
img=cv2.merge((b,g,r))
#b=img[:,:,0]
```

```python
img[:,:,2]=0
```

```python
plt.imshow(img)
```

## Making Borders for Images (Padding)

```python
BLUE=[255,0,0]
img1=cv2.imread('opencv_logo.png')
replicate=cv2.copyMakeBorder(img1,100,100,100,100,cv2.BORDER_REPLICATE)
reflect=cv2.copyMakeBorder(img1,20,20,20,20,cv2.BORDER_REFLECT)
reflect101=cv2.copyMakeBorder(img1,20,20,20,20,cv2.BORDER_REFLECT_101)
wrap=cv2.copyMakeBorder(img1,20,20,20,20,cv2.BORDER_WRAP)
constant=cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_CONSTANT,value=BLUE)
plt.subplot(231),plt.imshow(img1,'gray'),plt.title('ORIGINAL', fontsize=50)
plt.subplot(232),plt.imshow(replicate,'gray'),plt.title('REPLICATE', fontsize=50)
plt.subplot(233),plt.imshow(reflect,'gray'),plt.title('REFLECT', fontsize=50)
plt.subplot(234),plt.imshow(reflect101,'gray'),plt.title('REFLECT_101', fontsize=50)
plt.subplot(235),plt.imshow(wrap,'gray'),plt.title('WRAP', fontsize=50)
plt.subplot(236),plt.imshow(constant,'gray'),plt.title('CONSTANT', fontsize=50)
plt.rcParams["figure.figsize"] = (30,20)
plt.show()
```
