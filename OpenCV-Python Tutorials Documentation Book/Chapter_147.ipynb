# Canny Edge Detection

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('messi.jpg')
```

```python
edges1 = cv2.Canny(img, 200, 135)
```

```python
blur=cv2.GaussianBlur(img,(9, 9),0)
edges2 = cv2.Canny(blur, 200, 135)
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(131),plt.imshow(img, 'gray')
plt.title('Original Image', fontsize=25)
plt.subplot(132),plt.imshow(edges1, 'gray')
plt.title('Edge Image', fontsize=25)
plt.subplot(133),plt.imshow(edges2, 'gray')
plt.title('Edge Image \n& Gaussian Blur', fontsize=25)
plt.show()
```
