# Image Pyramids

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img=cv2.imread('messi.jpg')
```

```python
l=cv2.pyrDown(img)
l=cv2.pyrDown(l)
l=cv2.pyrDown(l)

h=cv2.pyrUp(l)
h=cv2.pyrUp(h)
h=cv2.pyrUp(h)

plt.figure(figsize=(15, 10))
plt.subplot(131),plt.imshow(img, 'gray')
plt.title('Original Image', fontsize=25)
plt.subplot(132),plt.imshow(l, 'gray')
plt.title('Lower Resolution', fontsize=25)
plt.subplot(133),plt.imshow(h, 'gray')
plt.title('Higher Resolution\n(final)', fontsize=25)
plt.show()
```
