# Smoothing Images

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## 2D Convolution ( Image Filtering )

```python
img=cv2.imread('r.png')

kernel = np.ones((5,5),np.float32)/25
dst=cv2.filter2D(img,-1,kernel)

plt.figure(figsize=(15, 10))
plt.subplot(121),plt.imshow(img),plt.title('Original', fontsize=25)
plt.subplot(122),plt.imshow(dst),plt.title('Averaging', fontsize=25)
plt.show()
```

## Averaging

```python
img=cv2.imread('r.png')

blur = cv2.blur(img, (5,5))

plt.figure(figsize=(15, 10))
plt.subplot(121),plt.imshow(img),plt.title('Original', fontsize=25)
plt.subplot(122),plt.imshow(blur),plt.title('Blur', fontsize=25)
plt.show()
```

## Gaussian Filtering

```python
img=cv2.imread('r.png')

blur = cv2.GaussianBlur(img,(5,5),0)

plt.figure(figsize=(15, 10))
plt.subplot(121),plt.imshow(img),plt.title('Original', fontsize=25)
plt.subplot(122),plt.imshow(blur),plt.title('Blur', fontsize=25)
plt.show()
```

### Class

```python
img=cv2.imread('blur2.jpg')

blur = cv2.GaussianBlur(img,(7,7),0)

plt.figure(figsize=(15, 10))
plt.subplot(121),plt.imshow(img),plt.title('Original', fontsize=25)
plt.subplot(122),plt.imshow(blur),plt.title('Blur', fontsize=25)
plt.show()
```

## Median Filtering

```python
img=cv2.imread('r.png')

median = cv2.medianBlur(img,5)

plt.figure(figsize=(15, 10))
plt.subplot(121),plt.imshow(img),plt.title('Original', fontsize=25)
plt.subplot(122),plt.imshow(median),plt.title('median', fontsize=25)
plt.show()
```

### Class

```python
img=cv2.imread('blur1.png')

median = cv2.medianBlur(img,5)

plt.figure(figsize=(15, 10))
plt.subplot(121),plt.imshow(img),plt.title('Original', fontsize=25)
plt.subplot(122),plt.imshow(median),plt.title('Blur', fontsize=25)
plt.show()
```

## Bilateral Filtering

```python
img=cv2.imread('r.png')

blur = cv2.bilateralFilter(img,9,75,75)

plt.figure(figsize=(15, 10))
plt.subplot(121),plt.imshow(img),plt.title('Original', fontsize=25)
plt.subplot(122),plt.imshow(blur),plt.title('median', fontsize=25)
plt.show()
```
