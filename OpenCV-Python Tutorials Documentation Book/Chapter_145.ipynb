```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img=cv2.imread('j.png',0)
kernel=np.ones((9,9),np.uint8)
erosion=cv2.erode(img,kernel,iterations=1)
dilation=cv2.dilate(img,kernel,iterations=1)
opening=cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
closing=cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
gradient=cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
tophat=cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)
blackhat=cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(241),plt.imshow(img),plt.title('Original', fontsize=25)
plt.subplot(242),plt.imshow(erosion),plt.title('Erosion', fontsize=25)
plt.subplot(243),plt.imshow(dilation),plt.title('Dilation', fontsize=25)
plt.subplot(244),plt.imshow(opening),plt.title('Opening', fontsize=25)
plt.subplot(245),plt.imshow(closing),plt.title('Closing', fontsize=25)
plt.subplot(246),plt.imshow(gradient),plt.title('Morphological\nGradient', fontsize=25)
plt.subplot(247),plt.imshow(tophat),plt.title('Top Hat', fontsize=25)
plt.subplot(248),plt.imshow(blackhat),plt.title('Black Hat', fontsize=25)
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(241),plt.imshow(img, 'gray'),plt.title('Original', fontsize=25)
plt.subplot(242),plt.imshow(erosion, 'gray'),plt.title('Erosion', fontsize=25)
plt.subplot(243),plt.imshow(dilation, 'gray'),plt.title('Dilation', fontsize=25)
plt.subplot(244),plt.imshow(opening, 'gray'),plt.title('Opening', fontsize=25)
plt.subplot(245),plt.imshow(closing, 'gray'),plt.title('Closing', fontsize=25)
plt.subplot(246),plt.imshow(gradient, 'gray'),plt.title('Morphological\nGradient', fontsize=25)
plt.subplot(247),plt.imshow(tophat, 'gray'),plt.title('Top Hat', fontsize=25)
plt.subplot(248),plt.imshow(blackhat, 'gray'),plt.title('Black Hat', fontsize=25)
plt.show()
```

```python
img=cv2.imread('fp1.jpg', cv2.COLOR_BGR2HSV)
lower_color=np.array([0,0,50])
upper_color=np.array([360,10,100])

# Threshold the HSV image to get only blue colors
mask=cv2.inRange(img, lower_color, upper_color)

res=cv2.bitwise_and(img,img, mask=mask)
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(241),plt.imshow(img, 'gray'),plt.title('Original', fontsize=25)
plt.subplot(242),plt.imshow(res, 'gray'),plt.title('Result', fontsize=25)
```
